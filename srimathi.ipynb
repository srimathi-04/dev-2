```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("winequalityN.csv")
```

```python
df.tail()
```

```python
df.hist()
plt.show()
```

```python
plt.scatter(df['fixed acidity'],df['citric acid'])
plt.show()
```

```python
df.shape
```

```python
x=df.pH
y=df.quality
plt.scatter(x,y,color='brown')
plt.title("WINE QUALITY")
plt.xlabel("pH")
plt.ylabel("quality")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color="red")
plt.title("WINE QUALITY")
plt.xlabel("pH")
plt.ylabel("quality")
plt.legend()
plt.show()
```

```python
df.boxplot(grid=True)
```

```python
plt.pie(y,x)
plt.title("WINE QUALITY")
plt.show()
```

```python
plt.bar(x,y,label="BARCHART",color="pink")
plt.title("WINE QUALITY")
plt.xlabel("pH")
plt.ylabel("quality")
plt.legend()
plt.show()
```
